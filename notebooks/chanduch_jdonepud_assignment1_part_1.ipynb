### Handling mismached and null data

```python
print(penguins_df2.species.unique())
print(penguins_df2.gender.unique())
print(penguins_df2.island.unique())
```

```python
island_dict = {'Torgersen':'torgersen','torgersen':'torgersen','Biscoe':'biscoe','biscoe':'biscoe','Dream':'dream','dream':'dream','DREAM':'dream'}
species_dict ={'Adelie': 'adelie', 'adelie': 'adelie','Gentoo' :'gentoo', "GENTOO":'gentoo', 'gentoo' :'gentoo','CHINSTRAP': 'chinstrap', 'Chinstrap': 'chinstrap', 'chinstrap':'chinstrap'}
gender_dict = {'MALE':'male','male':'male','female':'female', 'FEMALE':'female'}
penguins_df2.species = penguins_df2.species.map(species_dict)
penguins_df2.island = penguins_df2.island.map(island_dict)
penguins_df2.gender = penguins_df2.gender.map(gender_dict)
```

### Part1 Diamond data set

```python
diamond_df = pd.read_csv('diamond.csv')
```

```python
diamond_df
```

```python
diamond_df2 = diamond_df.copy()
```

```python
diamond_df2.info()
```

```python
diamond_df2.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
diamond_df2.carat.unique()
```

```python
diamond_df2 = diamond_df2[(diamond_df2.carat != 'TRUE') & (diamond_df2.carat != 'FALSE')]
diamond_df2 = diamond_df2[diamond_df2.carat != 'MAYBE']
diamond_df2 = diamond_df2.astype({'carat': float})
```

```python
diamond_df2 = diamond_df2[(diamond_df2.depth != 'MAYBE')&((diamond_df2.depth != 'FALSE') &(diamond_df2.depth != 'TRUE'))]
diamond_df2 = diamond_df2.astype({'depth': float})
```

```python
diamond_df2 = diamond_df2[(diamond_df2.table != 'MAYBE')&((diamond_df2.table != 'FALSE') &(diamond_df2.table != 'TRUE'))]
diamond_df2 = diamond_df2.astype({'table': float})
```

```python
diamond_df2.price.unique()
```

```python
diamond_df2 = diamond_df2[(diamond_df2.price != 'MAYBE')&((diamond_df2.price != 'FALSE') &(diamond_df2.price != 'TRUE'))]
diamond_df2 = diamond_df2.astype({'price': float})
```

```python
diamond_df2 = diamond_df2[(diamond_df2.x != 'MAYBE')&((diamond_df2.x != 'FALSE') &(diamond_df2.x != 'TRUE'))]
diamond_df2 = diamond_df2.astype({'x': float})
```

```python
diamond_df2 = diamond_df2[(diamond_df2.y != 'MAYBE')&((diamond_df2.y != 'FALSE') &(diamond_df2.y != 'TRUE'))]
diamond_df2 = diamond_df2[(diamond_df2.z != 'MAYBE')&((diamond_df2.z != 'FALSE') &(diamond_df2.z != 'TRUE'))]
diamond_df2 = diamond_df2.astype({'y': float})
diamond_df2 = diamond_df2.astype({'z': float})
```

```python
diamond_df2.describe()
```

```python
diamond_df2 = diamond_df2[(diamond_df2.cut != 'MAYBE')&((diamond_df2.cut != 'FALSE') &(diamond_df2.cut != 'TRUE'))]
diamond_df2 = diamond_df2[(diamond_df2.color != 'MAYBE')&((diamond_df2.color != 'FALSE') &(diamond_df2.color != 'TRUE'))]
diamond_df2 = diamond_df2[(diamond_df2.clarity != 'MAYBE')&((diamond_df2.clarity != 'FALSE') &(diamond_df2.clarity != 'TRUE'))]
```

```python
print(diamond_df2.cut.unique())
print(diamond_df2.color.unique())
print(diamond_df2.clarity.unique())
```

```python
cut_dict= {
    'Ideal':'ideal', 'IDEAL': 'ideal', 'ideal': 'ideal',
    'Good': 'good', 'GOOD':'good', 'good': 'good',
    'Fair': 'fair', 'FAIR':'fair', 'fair': 'fair',
    'Very Good': 'very_good', 'VERY GOOD' :'very_good', 'very good': 'very_good',
    'PREMIUM': 'premium', 'Premium': 'premium', 'premium': 'premium'}
color_dict = {
    'E':'e', 'I':'i','J':'j','H':'h','F':'f','G':'g', 'D':'d',
    'j':'j', 'd':'d', 'h':'h', 'g':'g', 'i':'i', 'f':'f', 'e':'e'
}
clarity_dict = {
    'SI2':'SI2', 'SI1':'SI1', 'VS1':'VS1', 'VS2':'VS2', 'VVS2':'VVS2', 'VVS1':'VVS1', 'I1':'I1', 'IF':'IF',
    'si1':'SI1', 'si2':'SI2', 'vs1':'VS1', 'vs2':'VS2', 'vvs2':'VVS2'
}
```

```python
diamond_df2.cut = diamond_df2.cut.map(cut_dict)
diamond_df2.color = diamond_df2.color.map(color_dict)
diamond_df2.clarity = diamond_df2.clarity.map(clarity_dict)

print(diamond_df2.cut.unique())
print(diamond_df2.color.unique())
print(diamond_df2.clarity.unique())
```
